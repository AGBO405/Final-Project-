## Housing Metric

To get the housing quality of a neighborhood, I created a "housing score". This score is equal to % of occupied units - % of vacant units - % of foreclosed units.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read in the data set
df = pd.read_csv("housing.csv")

# Convert percentage columns to numeric values
df['% Occupied Units (2010)'] = df['% Occupied Units (2010)'].str.replace('%', '').astype(float)
df['% Vacant Units (2010)'] = df['% Vacant Units (2010)'].str.replace('%', '').astype(float)
df['% of all Housing Units Foreclosed (2010)'] = df['% of all Housing Units Foreclosed (2010)'].str.replace('%', '').astype(float)
df['Population (2010)'] = df['Population (2010)'].str.replace(',', '').astype(float)

# Create the Metric
df = df[df['Population (2010)'] >= 500]
df['Housing Score'] = (df['% Occupied Units (2010)'] - df['% Vacant Units (2010)'] - df['% of all Housing Units Foreclosed (2010)'])

# Find the best neighborhood
best = df.sort_values(by='Housing Score', ascending=False)

# Visualize the data
top10 = best.head(20)
plt.figure(figsize=(10, 8))
plt.barh(top10['Neighborhood'], top10['Housing Score'])
plt.xlabel("Housing Score")
plt.title("Top 10 Neighborhoods by Housing Score")
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.gca().invert_yaxis()
plt.show()
```

## Analysis

What I found to be the best neighborhood was **Swisshelm Park**, as it had the highest housing score. This was mostly down to its large percentage of occupied units and its low % of vacant units and foreclosed units.

## Conclusion

Housing data helps identify stable neighborhoods by highlighting occupancy, vacancy, and foreclosure trends.
