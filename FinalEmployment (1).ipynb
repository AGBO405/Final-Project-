## Introduction

For my section of the project, I focused on economic opportunity. Using employment data, I analyzed how many residents in each neighborhood are employed. Higher employment rates suggest stronger economic stability and opportunity within a neighborhood.

## Employment Metric

To measure economic opportunity, I created an "employment score". This score is equal to the number of employed residents divided by the total population of each neighborhood.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read in the data set
df = pd.read_csv("employment.csv")

# Convert columns to numeric values
df["Population (2010)"] = df["Population (2010)"].str.replace(",", "").astype(float)
df["Total Age 16+ N'hood Residents Employed (2010 Est.)"] = pd.to_numeric(
    df["Total Age 16+ N'hood Residents Employed (2010 Est.)"], errors="coerce"
)

# Create the Metric
df = df[df["Population (2010)"] >= 500]

df["Employment Score"] = (
    df["Total Age 16+ N'hood Residents Employed (2010 Est.)"] /
    df["Population (2010)"]
)

# Find the best neighborhood
best = df.sort_values(by="Employment Score", ascending=False)

# Visualize the data
top10 = best.head(10)

plt.figure(figsize=(10, 8))
plt.barh(top10["Neighborhood"], top10["Employment Score"])
plt.xlabel("Employment Score")
plt.title("Top 10 Neighborhoods by Employment Score")
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.gca().invert_yaxis()
plt.show()
```

## Analysis

What I found to be the best neighborhood was **Spring Garden**, as it had the highest employment score. This is likely due to a higher proportion of residents being employed compared to other neighborhoods.

## Conclusion

Employment data helps identify economically strong neighborhoods by highlighting workforce participation. After filtering for populations over 500, the results are more reliable and less influenced by small sample sizes. However, employment alone does not fully determine the "best" neighborhood, as other factors like housing and safety also play a role.
