```python
# Economic Opportunity in Pittsburgh Neighborhoods

## Introduction
For my part of the project, I analyzed economic opportunity across Pittsburgh neighborhoods using the Employment 2010 dataset.

I define economic opportunity based on:
- Job availability (job density)
- Access to high-skill industries
- Employment stability

This helps identify which neighborhoods provide the strongest economic advantages.
```

```python
## Dataset

The dataset used is the Employment 2010 dataset from the City of Pittsburgh SNAP Census Data.

This dataset includes:
- Population of each neighborhood
- Number of jobs located in each neighborhood
- Distribution of employment across industries

This allows us to analyze both job availability and workforce composition.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/employment.csv")
df.columns = df.columns.str.strip()
```

```python
percent_cols = [
    "Resident Jobs: Prof, scientific, admin, and waste mgmt",
    "Resident Jobs: Finance, insurance, real estate",
    "Resident Jobs: Information",
    "Resident Jobs: Educ, health and social svc"
]

for col in percent_cols:
    df[col] = df[col].str.replace("%", "", regex=False)
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
df["job_density"] = df["Total # Jobs Located in N'hood (2000 - latest yr. avail.)"] / df["Population (2010)"]
```

```python
df["high_skill"] = (
    df["Resident Jobs: Prof, scientific, admin, and waste mgmt"] +
    df["Resident Jobs: Finance, insurance, real estate"] +
    df["Resident Jobs: Information"]
)

df["stability"] = df["Resident Jobs: Educ, health and social svc"]
```

```python
def normalize(col):
    return (col - col.min()) / (col.max() - col.min())

df["job_density_norm"] = normalize(df["job_density"])
df["high_skill_norm"] = normalize(df["high_skill"])
df["stability_norm"] = normalize(df["stability"])
```

```python
df["economic_score"] = (
    df["job_density_norm"] +
    df["high_skill_norm"] +
    df["stability_norm"]
) / 3
```

```python
top10 = df.sort_values("economic_score", ascending=False).head(10)
top10[["Neighborhood", "economic_score"]]
```

```python
top10.plot(
    x="Neighborhood",
    y="economic_score",
    kind="bar",
    title="Top 10 Neighborhoods by Economic Opportunity",
    legend=False
)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
## Analysis

The results show that neighborhoods with a high number of jobs and strong representation in professional, finance, and healthcare industries tend to rank highest.

This suggests that economic opportunity is influenced not only by job availability but also by the type and stability of employment sectors.
```

```python
## Conclusion

Based on this analysis, the best neighborhoods in terms of economic opportunity are those that combine high job density with strong high-skill and stable industries.

This provides a data-driven perspective on what makes a neighborhood “best” from an economic standpoint.
```
