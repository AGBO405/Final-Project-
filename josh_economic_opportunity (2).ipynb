# Introduction

For my section of the project, I focused on economic opportunity. Using employment data, I determined the best neighborhood by comparing employment rates across Pittsburgh neighborhoods with a minimum population of 500 people. This helps show which neighborhoods have stronger access to employment and greater economic stability.

# Economic Opportunity Metric

To measure economic opportunity in each neighborhood, I created an "employment rate." This rate is equal to the total number of employed residents divided by the total population. I also filtered out neighborhoods with populations below 500 so that very small neighborhoods would not distort the results.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read in the data set
df = pd.read_csv("employment.csv", low_memory=False)

# Select needed columns
df = df[[
    "Neighborhood",
    "Population (2010)",
    "Total Age 16+ N'hood Residents Employed (2010 Est.)"
]]

# Rename columns
df.columns = ["Neighborhood", "Population", "Employed"]

# Clean the data
df = df.dropna()

# Create the metric
df = df[df["Population"] >= 500]
df["Employment Rate"] = df["Employed"] / df["Population"]

# Find the best neighborhood
best = df.sort_values(by="Employment Rate", ascending=False)

# Visualize the data
top10 = best.head(10)
plt.figure(figsize=(10, 8))
plt.barh(top10["Neighborhood"], top10["Employment Rate"])
plt.xlabel("Employment Rate")
plt.title("Top 10 Neighborhoods by Employment Rate")
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.gca().invert_yaxis()
plt.show()
```

```python
# Analysis

What I found to be the best neighborhood was **New Homestead**, as it had the highest employment rate. This was mostly due to its high number of employed residents relative to its total population, which suggests a strong level of economic opportunity compared to other neighborhoods.
```

# Conclusion

Employment data helps identify neighborhoods with stronger economic opportunity by showing where a larger share of residents are employed. After filtering for populations over 500, the results are more reliable and less affected by very small sample sizes.
